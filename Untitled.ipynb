```python
!pip install basemap
!pip install folium 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import folium
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster


# Load the datasets
nashville_data = pd.read_csv('resources/listings (1).csv')
nyc_data = pd.read_csv('resources/listings (2).csv')

# Display the first few rows of each dataset and their columns
nashville_head = nashville_data.head()
nyc_head = nyc_data.head()
nashville_columns = nashville_data.columns
nyc_columns = nyc_data.columns

nashville_head, nyc_head, nashville_columns, nyc_columns
```

```python
#clean data 
# Drop 'neighbourhood_group' column from Nashville dataset
nashville_data.drop('neighbourhood_group', axis=1, inplace=True)

# Handle missing values for both datasets

# Nashville
nashville_data['last_review'].fillna('No reviews', inplace=True)
nashville_data['reviews_per_month'].fillna(0, inplace=True)

# NYC
nyc_data['host_name'].fillna('Unknown', inplace=True)
nyc_data['last_review'].fillna('No reviews', inplace=True)
nyc_data['reviews_per_month'].fillna(0, inplace=True)

# Drop 'license' column from both datasets due to large number of missing values
nashville_data.drop('license', axis=1, inplace=True)
nyc_data.drop('license', axis=1, inplace=True)

# Check if there are any more missing values left
missing_values_nashville_cleaned = nashville_data.isnull().sum().max()
missing_values_nyc_cleaned = nyc_data.isnull().sum().max()

missing_values_nashville_cleaned, missing_values_nyc_cleaned
```

```python
# Calculate average total vacancy for AirBnbs in each city
nyc_avg_availability = nyc_data['availability_365'].mean()
nashville_avg_availability = nashville_data['availability_365'].mean()

# Calculate average availability for each neighborhood in NYC and Nashville
nyc_neighbourhood_avg_availability = nyc_data.groupby('neighbourhood')['availability_365'].mean().sort_values(ascending=False)
nashville_neighbourhood_avg_availability = nashville_data.groupby('neighbourhood')['availability_365'].mean().sort_values(ascending=False)

nyc_avg_availability, nashville_avg_availability, nyc_neighbourhood_avg_availability, nashville_neighbourhood_avg_availability
```

```python
# Average availability for Nashville and NYC
nashville_avg_availability = nashville_data['availability_365'].mean()
nyc_avg_availability = nyc_data['availability_365'].mean()

# Average availability by neighbourhood for Nashville and NYC
nashville_neighbourhood_avg_availability = nashville_data.groupby('neighbourhood')['availability_365'].mean().sort_values()
nyc_neighbourhood_avg_availability = nyc_data.groupby('neighbourhood')['availability_365'].mean().sort_values()

# Visualization 1: Bar chart comparing average availability between NYC and Nashville
cities = ['Nashville', 'NYC']  # Switched order of cities
avg_availabilities = [nashville_avg_availability, nyc_avg_availability]  # Switched order of avg_availabilities

plt.figure(figsize=(8, 6))
plt.bar(cities, avg_availabilities, color=['steelblue', 'green'])  # Nashville in blue, NYC in green
plt.title('Average Vacant Days per Year: Nashville vs NYC')  # Switched order in title
plt.ylabel('Average Days Available')
plt.show()

# Visualization 2: Bar chart for top 10 neighborhoods with lowest average availability in both cities
combined_neighbourhood_least_availability = pd.concat([
    nashville_neighbourhood_avg_availability.head(5),  # Nashville data first
    nyc_neighbourhood_avg_availability.head(5)  # NYC data next
])

plt.figure(figsize=(16, 8))
combined_neighbourhood_least_availability.plot(kind='bar', color=['steelblue']*5 + ['green']*5)  # Colors remain the same
plt.title('Average Vacant Days per Year by Neighborhood')
plt.ylabel('Average Days Vacant')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
# Plot the Nashville data on the first subplot
sns.boxplot(y=nashville_neighbourhood_avg_availability, ax=axes[0], palette='Blues_d')
axes[0].set_title('Nashville Availability')
axes[0].set_ylabel('Availability 365')

# Plot the NYC data on the second subplot
sns.boxplot(y=nyc_neighbourhood_avg_availability, ax=axes[1], palette='Greens_d')
axes[1].set_title('NYC Availability')
axes[1].set_ylabel('Availability 365')

# Set a main title for the whole figure
fig.suptitle('Days unit is vacant per year', fontsize=15)

plt.tight_layout()
plt.show()
```
