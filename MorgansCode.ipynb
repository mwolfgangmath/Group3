```python
import matplotlib.pyplot as plt

cities = ['Nashville', 'New York']
tourists = [14400000, 56400000]  # Replace these numbers with actual data

plt.bar(cities, tourists, color=['blue', 'green'])
plt.xlabel('Cities')
plt.ylabel('Number of Tourists')
plt.title('Number of Tourists in Nashville and New York')
plt.show()
```
